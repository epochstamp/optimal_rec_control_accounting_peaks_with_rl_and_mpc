```python
import pandas as pd
import os
```

```python
root_dir = os.path.expanduser('~') + "/OneDrive/wp1/"
ymr = pd.read_parquet(root_dir + "data/YMRs/ymr.parquet")
amr = pd.read_parquet(root_dir + "data/AMRs/amr.parquet")
```

```python
ymr.index = amr.index
```

```python
df = pd.concat([ymr, amr], axis=1)
```

```python
df = df.groupby(pd.Grouper(freq='15min')).sum()
```

```python
df.to_parquet(root_dir + "data/full_mrs/meterings.parquet")
```
